```python
sum_array = test['value'].to_numpy()
```

Create a dictionary with the main MK output parameters (existence of a trend, slope) as keys and their corresponding function output values per each cell as values. Here we should describe each of the MK parameters somehow and say what they show and that we use them to plot later.

```python
array_dict = {}
array_dict['trend'] = []
array_dict['slope'] = []
for a in sum_array:
    result = mk.original_test(a)
    array_dict['trend'].append(result.trend)
    array_dict['slope'].append(result.slope)
```

```python
test['trend'] = array_dict['trend']
test['slope'] = array_dict['slope']
display(test['trend'].value_counts())
display(test['slope'].describe())
display(test.head())
```

```python
test['trend_num'] = test['trend'].apply(lambda x: -1 if x == 'decreasing' else (1 if x == 'increasing' else 0))
test
```

```python
months = test['month'].unique()
for month in months:
    subset = test[test['month'] == month]
    precip_ds['trend' + month] = subset['trend_num']
```
